```python
import damask
import numpy as np
```

```python
rnd = damask.Rotation.from_random(82*2)
rnd1 = damask.Rotation.from_random(18*2)
```

```python
config = damask.ConfigMaterial()\
               .material_add(O=rnd,phase='Cu',homogenization='SX')\
               .material_add(O=rnd1,phase='Nb',homogenization='SX')\
              # .material_add(O=sph,phase='phase_B',homogenization='SX')
```

```python
print(f'configuration is{" " if config.is_valid else " not "}valid\n')
len(config['material'])
```

```python
config.save()
config
```
